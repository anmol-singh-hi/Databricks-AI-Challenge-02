```python
import pandas as pd
import datetime

df=pd.DataFrame({
    'birth_year': [1990, 1985, 2000, 1975, 1995],
    'income': [25000, 45000, 80000, 32000, 70000],
    'last_login': ['2026-02-10', '2026-01-15', '2026-02-20', '2025-12-25', '2026-02-01']
})
display(df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['birth_year'], df['income'])
plt.xlabel('birth_year')
plt.ylabel('income')
plt.title('Scatter Plot of birth_year vs income')
plt.show()
```
