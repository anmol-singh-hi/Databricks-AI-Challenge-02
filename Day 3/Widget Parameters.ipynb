# What do you mean by Parameterized Notebooks?
Parameterized notebooks in Databricks allow you to pass input parameters to a notebook, making your workflows more flexible and reusable. This is especially useful for orchestrating notebook workflows, modularizing code, and automating tasks. The most common way to parameterize notebooks is by using Databricks widgets or by passing arguments when invoking a notebook from another notebook using the dbutils.notebook.run() method.

```python
dbutils.widgets.text("widet_name", defaultValue="this is default value", label="this is label")
```

```python
dbutils.widgets.get("widet_name")
```

```python
dbutils.widgets.text("first", defaultValue="0", label="Enter first integer")
dbutils.widgets.text("second", defaultValue="0", label="Enter second integer")
```

```python
first = int(dbutils.widgets.get("first"))
second = int(dbutils.widgets.get("second"))
dbutils.widgets.dropdown("operation", "add", ["add", "subtract", "multiply", "divide"], "Choose operation")
operation = dbutils.widgets.get("operation")

if operation == "add":
    result = first + second
elif operation == "subtract":
    result = first - second
elif operation == "multiply":
    result = first * second
elif operation == "divide":
    result = first / second if second != 0 else None
else:
    result = None

display(result)
```
