```python
df = spark.read.csv("/Volumes/luffy/phase2/bronze/data.csv")
display(df.head(10))
```

```python
df.count(),len(df.columns)
```

```python
# Extract first row as column names
first_row = df.limit(1).collect()[0]
new_columns = [str(item) for item in first_row]

# Remove the first row from the DataFrame using DataFrame API
df_data = df.filter(df['_c0'] != first_row[0])

# Rename columns
df_data = df_data.toDF(*new_columns)

display(df_data.head(10))
```

```python
df_data.count(),len(df_data.columns)
```

```python
df_data.write.format("delta").mode("overwrite").save("/Volumes/luffy/phase2/bronze/ecom_transactions")
```
